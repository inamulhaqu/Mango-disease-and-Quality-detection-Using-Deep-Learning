```python
import tensorflow as tf
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "mango_disease_dataset",
    shuffle=True,
     image_size=(224,224)
)
```

```python
  # Adjust the image size as needed
resize_and_rescale = tf.keras.Sequential([
    layers.experimental.preprocessing.Resizing(250, 250)
    
])
```

```python
data_augmentation = tf.keras.Sequential([
  layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
  layers.experimental.preprocessing.RandomRotation(0.2),
])
```

```python
for image_batch, label_batch in dataset.take(1):
    print(image_batch.shape)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# Assuming you have a dataset containing images and labels
# Replace 'dataset' and 'class_names' with your actual dataset and class names
for image_batch, label_batch in dataset.take(1):
    for i in range(4):
        original_image = image_batch[i].numpy().astype("uint8")
        preprocessed_image = resize_and_rescale(image_batch[i:i+1]).numpy().astype("uint8")

        plt.figure(figsize=(10, 10))
         
        # Display the original image
        plt.subplot(1, 2, 1)
        plt.imshow(original_image)
        plt.title("Original Image")
        plt.axis("off")

        # Display the preprocessed image
        plt.subplot(1, 2, 2)
        plt.imshow(preprocessed_image[0])
        plt.title("Resized and Rescaled Image")
        plt.axis("off")

        plt.show()
```

```python

for image_batch, label_batch in dataset.take(1):
    for i in range(4):
        if i < len(image_batch):
            original_image = image_batch[i].numpy().astype("uint8")
            
            # Apply data augmentation
            augmented_image = data_augmentation(image_batch[i:i+1])[0].numpy().astype("uint8")

            plt.figure(figsize=(10, 5))

            # Display the original image with size information
            plt.subplot(1, 2, 1)
            plt.imshow(original_image)
            plt.title(f"Original Image\nSize: {original_image.shape[0]}x{original_image.shape[1]}")
            plt.axis("off")

            # Display the augmented image with size information
            plt.subplot(1, 2, 2)
            plt.imshow(augmented_image)
            plt.title(f"Augmented Image\nSize: {augmented_image.shape[1]}x{augmented_image.shape[0]}\nData Augmentation: Flip & Rotate")
            plt.axis("off")

            plt.show()
        else:
            break




```
